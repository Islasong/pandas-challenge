```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
school_data_complete = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(school_data_complete, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
#Calaulate the total number of students
student_count = len(school_data_complete["Student ID"].unique())
student_count
```

```python
#Calaulate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calaulate the average maths score 
average_maths = school_data_complete["maths_score"].mean()
average_maths
```

```python
#Calaulate the average reading score
average_reading = school_data_complete["reading_score"].mean()
average_reading
```

```python
#Calaulate %passing maths(50 or higher)
student_pass_maths = school_data_complete.loc[school_data_complete["maths_score"]>=50]
student_pass_maths_count = student_pass_maths["Student ID"].count()
percentage_passing_maths = (student_pass_maths_count / student_count) *100
percentage_passing_maths
```

```python
#Calculate %passing reading
student_pass_reading = school_data_complete.loc[school_data_complete["reading_score"]>=50]
student_pass_reading_count = student_pass_reading["Student ID"].count()
percentage_passing_reading = (student_pass_reading_count / student_count) *100
percentage_passing_reading
```

```python
#Calculate %overall passing
student_pass_overall = school_data_complete.loc[(school_data_complete["reading_score"]>=50) & (school_data_complete["maths_score"]>=50)]
student_pass_overall_count = student_pass_overall["Student ID"].count()
percentage_passing_overall = (student_pass_overall_count / student_count) *100
percentage_passing_overall
```

```python
#Create DataFrame
summary_df = pd.DataFrame(
    {"Total Schools": school_count,
     "Total Students": f"{student_count:,}",
     "Total Budget": f"${total_budget:,.2f}",
     "Average Maths Score": f"{average_maths:.6f}",
     "Average Reading Score": f"{average_reading:.6f}",
     "% Passing Maths": f"{percentage_passing_maths:.6f}",
     "% Passing Reading": f"{percentage_passing_reading:.6f}",
     "% Overall Passing": f"{percentage_passing_overall:.6f}"
     }, index=[0])

summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Use groupby for school name
school_name = school_data_complete.set_index("school_name").groupby(["school_name"])
school_type = school_data.set_index("school_name")["type"]
total_student = school_name["Student ID"].count()
total_school_budget = school_data.set_index('school_name')["budget"]
per_student_budget = school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"]
Average_maths_score = school_name["maths_score"].mean()
Average_reading_score = school_name["reading_score"].mean()
Percentage_Passing_maths = school_data_complete[school_data_complete["maths_score"]>=50].groupby(["school_name"])["Student ID"].count()/total_student*100
Percentage_Passing_reading = school_data_complete[school_data_complete["reading_score"]>=50].groupby(["school_name"])["Student ID"].count()/total_student*100
Percentage_Passing_overall = school_data_complete[(school_data_complete["reading_score"]>=50) & (school_data_complete["maths_score"]>=50)].groupby(["school_name"])["Student ID"].count()/total_student*100

#Create DataFrame for school summary
school_summary_df = pd.DataFrame(
    {"School Type": school_type,
     "Total Students": total_student,
     "Total School Budget": total_school_budget,
     "Per Student Budget": per_student_budget,
     "Average Maths Score": Average_maths_score,
     "Average Reading Score": Average_reading_score,
     "% Passing Maths": Percentage_Passing_maths,
     "% Passing Reading": Percentage_Passing_reading,
     "% Overall Passing": Percentage_Passing_overall
     })
school_summary_df
```

## Top Performing Schools (By % Overall Passing)

```python
#Sort and display the top five performing school by % overall passing.
top_performing_school = school_summary_df.sort_values("% Overall Passing", ascending=False)
top_performing_school.index.name = ("")
top_performing_school.head(5)
```

```python
#Year 9
maths_year9_df = school_data_complete.loc[school_data_complete["year"] == 9,["school_name", "maths_score"]]
maths_year9_score = maths_year9_df.groupby(["school_name"])
year9_average_math = maths_year9_score.mean()
year9_average_math = year9_average_math.rename(columns ={"maths_score":"Year 9"})
year9_average_math


```

```python
#Year 10
maths_year10_df = school_data_complete.loc[school_data_complete["year"] == 10,["school_name", "maths_score"]]
maths_year10_score = maths_year10_df.groupby(["school_name"])
year10_average_math = maths_year10_score.mean()
year10_average_math = year10_average_math.rename(columns ={"maths_score":"Year 10"})
year10_average_math
```

```python
# Year 11
maths_year11_df = school_data_complete.loc[school_data_complete["year"] == 11,["school_name", "maths_score"]]
maths_year11_score = maths_year11_df.groupby(["school_name"])
year11_average_math = maths_year11_score.mean()
year11_average_math = year11_average_math.rename(columns ={"maths_score":"Year 11"})
year11_average_math
```

```python
#Year 12
maths_year12_df = school_data_complete.loc[school_data_complete["year"] == 12,["school_name", "maths_score"]]
maths_year12_score = maths_year12_df.groupby(["school_name"])
year12_average_math = maths_year12_score.mean()
year12_average_math = year12_average_math.rename(columns ={"maths_score":"Year 12"})
year12_average_math
```

```python
#Merge them 
maths_score_by_year_df1 = pd.merge(year9_average_math,year10_average_math, on="school_name")
maths_score_by_year_df2 = pd.merge(year11_average_math,year12_average_math,on="school_name")
math_score_by_year = pd.merge(maths_score_by_year_df1,maths_score_by_year_df2, on="school_name")
math_score_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#The maximum value of per student budget
school_summary_df["Per Student Budget"].max()
```

```python
#The minimum value of per student budget
school_summary_df["Per Student Budget"].min()
```

```python
#Create bins
bins = [0, 585, 630, 645, 680]
budget_range = ["<$585", "$585-630", "$630-645","$645-680"]
school_summary_df ["Per Student Budget"]=pd.cut(school_summary_df ["Per Student Budget"], bins, labels=budget_range, include_lowest=True)
#Rename column head to Spending Ranges(Per Student)
school_summary_df = school_summary_df.rename(columns = {"Per Student Budget":"Spending Ranges (Per Student)"})

school_summary_df.head()
```

```python
#Create Grouby Spending Ranges (per student)
spending_group = school_summary_df.groupby(["Spending Ranges (Per Student)"])

spending_summary = spending_group[["Average Maths Score", 
                                   "Average Reading Score",
                                   "% Passing Maths",
                                   "% Passing Reading",
                                   "% Overall Passing"]].mean()
#formatting data by spending Ranges
spending_summary = spending_summary.style.format("{:.2f}")

spending_summary
```
